## Load Required Packages & Data

```python
# load python packages from environment
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
# Get current working directory
path = os.getcwd()

# Get parent directory
parent_path = os.path.dirname(path)
print(parent_path)

# Create data/raw path
interimdata_path = os.path.join(parent_path, 'data', 'interim')
print(interimdata_path)
```

```python
df = pd.read_csv(interimdata_path + '/1.0_cwb_pm25_pm10_clean1.csv')
df.head()
```

```python
# Fix dtypes
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

## Outliers and Anamolies

```python
df = df.set_index('date')
```

```python
# Visualize the monthly trend
fig, axes = plt.subplots(2,1)
fig.suptitle('Month-wise Box Plot 2014-2020\n(The Seasonality)')

sns.boxplot(ax=axes[0],x=df.index.month, y='pm25', data=df)
sns.boxplot(ax=axes[1],x=df.index.month, y='pm10', data=df)
axes[1].set_xlabel('Month')

fig.savefig(parent_path +'/visualizations/2.0_monthwise_boxplots_seasonality.jpg')
```

```python
# Visualize the monthly trend
fig, axes = plt.subplots(2,1)
fig.suptitle('Year-Wise Box Plot 2014-2020\n(The Trend)')

sns.boxplot(ax=axes[0],x=df.index.year, y='pm25', data=df)
sns.boxplot(ax=axes[1],x=df.index.year, y='pm10', data=df)
axes[1].set_xlabel('Year')

fig.savefig(parent_path +'/visualizations/2.0_yearwise_boxplots_trend.jpg')
```

## TimeSeries and Pollution Measurements

```python
from IPython.display import Image
aq_measure = parent_path + '/reference/air quality measurements.png'
Image(aq_measure)
```

```python
fig = df['pm25'].plot(figsize=(18,6),legend=False).get_figure()
plt.title('PM 2.5 Causeway Bay', size=18)

plt.fill_between(df.index,0,15, color='green', alpha=0.2) # Good
plt.fill_between(df.index,15.5,40.4, color='yellow', alpha=0.2) # Moderate
plt.fill_between(df.index,40.5, 65.4, color='orange', alpha=0.3) # Unhealthy for sensitive groups
plt.fill_between(df.index,65.5,150.4, color='red', alpha=0.2) # Unhealthy 
plt.fill_between(df.index,150.5,250.4, color='purple', alpha=0.2) # Very unhealthy
#plt.fill_between(df.index,250.5,500.4, color='brown', alpha=0.2) # Hazardous


plt.ylabel('PM 2.5', size=14)
plt.xlabel('Date', size=14)

fig.savefig(parent_path +'/visualizations/2.0_timeseries_pm25_color.jpg')
```

```python
fig = df['pm10'].plot(figsize=(18,6),legend=False).get_figure()
plt.title('PM 10 Causeway Bay', size=18)

plt.fill_between(df.index,0,54, color='green', alpha=0.2) # Good
plt.fill_between(df.index,55,154, color='yellow', alpha=0.2) # Moderate
#plt.fill_between(df.index,155,254, color='orange', alpha=0.3) # Unhealthy for sensitive groups
#plt.fill_between(df.index,255,354, color='red', alpha=0.2) # Unhealthy 
#plt.fill_between(df.index,355,424, color='purple', alpha=0.2) # Very unhealthy
#plt.fill_between(df.index,425,604, color='brown', alpha=0.2) # Hazardous

plt.ylabel('PM 10', size=14)
plt.xlabel('Date', size=14)

fig.savefig(parent_path +'/visualizations/2.0_timeseries_pm10_color.jpg')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(df.loc['2014']['pm25'].values)
ax.plot(df.loc['2015']['pm25'].values)
ax.plot(df.loc['2016']['pm25'].values)
ax.plot(df.loc['2017']['pm25'].values)
ax.plot(df.loc['2018']['pm25'].values)
ax.plot(df.loc['2019']['pm25'].values)
ax.plot(df.loc['2020']['pm25'].values)
plt.title('Monthly Trends (2014-2020)')
plt.legend(['2014','2015','2016','2017','2018','2019','2020'])
```
